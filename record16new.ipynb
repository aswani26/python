```python
class Bank:
    def __init__(self, acnumber, name, actype, balance):
        self.acnumber = acnumber
        self.name = name
        self.actype = actype
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Successfully deposited!")

    def withdraw(self, amount):
        if amount > self.balance:
            print(" Insufficient balance!")
        else:
            self.balance -= amount
            print(" Successfully withdraw amount!")

    def display(self):
        print("\n Account Details")
        print("Account Number:", self.acnumber)
        print("Name:", self.name)
        print("Account Type:", self.actype)
        print("Balance:", self.balance)


acnumber = int(input("Enter account number: "))
name = input("Enter name: ")
actype = input("Enter account type: ")
balance = float(input("Enter initial balance: "))

account = Bank(acnumber, name, actype, balance)

while (True):
    print("\n1. Deposit\n2. Withdraw\n3. Account Info\n4. Exit")
    ac = int(input("Enter your choice: "))

    if ac== 1:
        amount = float(input("Enter amount to deposit: "))
        account.deposit(amount)
    elif ac== 2:
        amount = float(input("Enter amount to withdraw: "))
        account.withdraw(amount)
    elif ac == 3:
        account.display()
    else:   
        print(" Exiting...")
        break
    
        
```
